```python
root_experiments = './experiments'
for name_model in  ['ae','ocsvm', 'deepsvdd', 'classvdd', 'iforest' ]:
    for hierClass in ['Transient','Periodic' ]:

        if hierClass=='Transient':
            possible_outliers = ['SLSN',
                                'SNII',
                                'SNIa',
                                'SNIbc']
        elif hierClass == 'Stochastic':
            possible_outliers = [ 'AGN' ,
                                'Blazar',
                                'CV/Nova',
                                'QSO',
                                'YSO']
        elif hierClass == 'Periodic':
            possible_outliers = ['CEP',
                                'DSCT',
                                'E',
                                'RRL', 
                                'LPV']

        auprs = []
        auprs_devs = []
        aurocs = []
        aurocs_devs = []

        for outlier in possible_outliers:
            auroc = []
            auroc_dev = []
            aupr = []
            aupr_dev = []
            for fold in range(5):
                fold = fold
                outlier = outlier
                
                job_name = '{}_{}_{}/fold{}'.format(name_model, hierClass, outlier,  fold)

                directory = os.path.join(root_experiments, job_name)

                data = json.load(open('{}/metrics_test.json'.format(directory)))
                auroc.append(data['AU ROC'])
                aupr.append(data['AU PR'])
            aurocs.append(np.mean(auroc))
            aurocs_devs.append(np.std(auroc))
            auprs.append(np.mean(aupr))
            auprs_devs.append(np.std(aupr))
            print(outlier, name_model )
            print('AU ROC {:.3f} +- {:.3f}'.format(aurocs[-1], aurocs_devs[-1]))
            print('AU PR {:.3f} +- {:.3f}'.format(auprs[-1], auprs_devs[-1]))
```
